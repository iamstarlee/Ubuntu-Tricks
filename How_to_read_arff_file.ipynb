- The arff file is coded by utf-8
- Names are after @attribute
- Data are after @data

```python
import pandas as pd

with open("xxx.arff", encoding="utf-8") as f:
    header = []
    for line in f:
        if line.startswith("@attribute"):
            header.append(line.split()[1])
        elif line.startswith("@data"):
            break
    df = pd.read_csv(f, header=None)
    df.columns = header
df
```
